## Global Superstore Orders

```python
import numpy as np
import pandas as pd

orders = pd.read_excel("C:\\ML\\dataset\\Global Superstore Orders 2016.xlsx")
orders.head()
```

```python
pd.crosstab(orders.Segment, orders.Category, values=orders.Profit, aggfunc='mean')
```

```python
pd.crosstab(orders.Segment, [orders.Category, orders.Market], values=orders.Profit, aggfunc='max' )
```

```python
import seaborn as sns

sns.heatmap(pd.crosstab([orders.Category, orders.Market], orders.Segment), cmap="YlGnBu", annot=True, cbar=True)
```

```python
pd.crosstab(orders.Market, orders.Category, normalize=True)
```

## Titanic

```python
import pandas as  pd
import numpy as np
import seaborn as sns

titanic = pd.read_excel('C:\\ML\\dataset\\titanic.xlsx')
titanic.head(10)
```

```python
titanic.groupby('sex')[['survived']].mean()
```

```python
titanic.groupby(['sex', 'pclass'])['survived'].aggregate('mean').unstack()
```

```python
titanic.pivot_table('survived', index='sex', columns='pclass')
```

```python
age = pd.cut(titanic['age'], [0, 18, 80])

titanic.pivot_table('survived', ['sex', age], 'pclass')
```

```python
fare = pd.qcut(titanic['fare'], 2)

titanic.pivot_table('survived', ['sex', age], [fare, 'pclass'])
```

```python
titanic.pivot_table(index='sex', columns='pclass', aggfunc={'survived':sum, 'fare':'mean'})
```

## Births

```python
import pandas as pd
birth= pd.read_csv('C:\\ML\\dataset\\births.csv')
birth.head(5)
```

```python
birth['decade'] = 10 * (birth['Year'] // 10)
birth.pivot_table('Live Births', index='decade', columns='Fertility Rate', aggfunc='sum')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

sns.set() # use Seaborn styles

birth.pivot_table('Live Births', index='decade', columns='Fertility Rate', aggfunc='sum').plot()
plt.ylabel('total births per year');
```
