```python
#مدریت دیتای طبقه بندی شده
import pandas as pd
shahin = pd.DataFrame([
                  ['green', 'M', 10.1, 'class1'],
                  ['red', 'L', 13.5, 'class2'],
                  ['blue', 'XL', 15.3, 'class1']])
shahin.columns = ['color', 'size', 'price', 'classlabel']
shahin
```

```python
#مپ کردن ویژگی های ترتیبی
# XL= L+1= M+2
size_mapping = {'XL': 3, 'L': 2, 'M': 1}
shahin['size'] = shahin['size'].map(size_mapping) 
shahin
                
               
```

```python
# برگرداندن مقادیر مپ شده به حالت قبل
inv_size_mapping = {v: k for k, v in size_mapping.items()}
shahin['size'] = shahin['size'].map(inv_size_mapping) 
shahin
```

```python
#Encoding class labels
import numpy as np
class_mapping = {label:idx for idx,label in enumerate(np.unique(shahin['classlabel']))}

class_mapping
```

```python
shahin['classlabel'] = shahin['classlabel'].map(class_mapping)
shahin
```

```python
# encoding on ویژگی های اسمی
from sklearn.preprocessing import LabelEncoder
X= shahin[['color', 'size', 'price']].values
color_le = LabelEncoder()
X[:, 0] = color_le.fit_transform(X[:, 0])
X
```

```python
columns=['color', 'size', 'price']
X= pd.DataFrame(data=X, index=None, columns=columns)
X
```

```python
Y=shahin['classlabel']
X= X.join(Y)
X

```

```python
#get dummies
shahin = pd.get_dummies(shahin[['price', 'color', 'size', 'classlabel']])
shahin
```
