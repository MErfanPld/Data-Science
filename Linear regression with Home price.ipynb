```python
import pandas as pd
import numpy as np
boston = pd.read_csv('C:\\ML\\dataset\\BostonHousing.csv')
boston.head()
```

```python
import statsmodels.api as sm # import statsmodels
```

```python
# Put the target (housing value -- MEDV) in another DataFrame
target = pd.DataFrame(data=boston.values[:, 13], columns=["MEDV"])
```

```python
target.head()
```

```python
#simple linear regression
import statsmodels.api as sm # import statsmodels 

X = boston["lstat"] ## X usually means our input variables (or independent variables)
y = target["MEDV"] ## Y usually means our output/dependent variable
X = sm.add_constant(X) ## let's add an intercept (beta_0) to our model

# Note the difference in argument order
model = sm.OLS(y, X).fit() ## sm.OLS(output, input)
predictions = model.predict(X)

# Print out the statistics
model.summary()

#
#
#
#  R-squared هرچقدر به یک نزدیک تر باشه بهتره 
```
